### 实现函数$p(x,\text{coeff})$，计算如下多项式。
\begin{equation}
p(x)=a_0+a_1x+a_2x^2+\dots+a_nx^n=\sum^n_{i=0}a_ix^i
\end{equation}
其中$a_i$由向量$\text{coeff}$提供。

```julia
p(x,coeff)=sum(a*x^(i-1) for (i,a) in enumerate(coeff))
x=2
coeff=[1,3,5,7,9]
p(x,coeff)
```

### 编写传入字符串为参数的函数，返回其中大写字符的个数。

```julia
function foo(string)
    count=0
    for letter in string
        if (letter==uppercase(letter)) && isletter(letter)
            count+=1
        end
    end
    return count
end

foo("Hello Yeruo,Hello World")
```

### 编写一个函数，两个序列$seq_a$、$seq_b$作为输入，如果所有$seq_a$中的元素也存在$seq_b$中，返回$true$，否则返回$false$。

```julia
function foo(seq_a,seq_b)
    for element in seq_a
        if element ∉ seq_b
            return false
        end
    end
    return true
end

println(foo([1, 2], [1, 2, 3]))
println(foo([1, 2, 3], [1, 2]))
```

```julia
function linapprox(f,a,b,x,n)
    len=b-a
    step=len/(n-1)
    start=a
    while start ≤ x
        start+=step
    end
    u,v=start-step,start
    return f(u)+(f(v)-f(u))/(v-u)*(x-u)
end

f(x)=sin(π*x)
foo(x)=linapprox(f,-5,5,x,20)
x=range(-5,5,length=100)
y=f.(x)
yy=foo.(x)
using Plots
plot(x,y,label="real line")
plot!(x,yy,label="linapprox line")
```
